### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Hola


#1. Análisis de facturación



```python
facturacion_tienda1 = tienda["Precio"].sum()
facturacion_tienda2 = tienda2["Precio"].sum()
facturacion_tienda3 = tienda3["Precio"].sum()
facturacion_tienda4 = tienda4["Precio"].sum()

print("Facturación Tienda 1:", facturacion_tienda1)
print("Facturación Tienda 2:", facturacion_tienda2)
print("Facturación Tienda 3:", facturacion_tienda3)
print("Facturación Tienda 4:", facturacion_tienda4)
```

```python
facturacion = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Facturacion": [
        facturacion_tienda1,
        facturacion_tienda2,
        facturacion_tienda3,
        facturacion_tienda4
    ]
}

df_facturacion = pd.DataFrame(facturacion)

df_facturacion
```

```python
import matplotlib.pyplot as plt

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

facturacion = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

plt.figure(figsize=(8,5))
plt.bar(tiendas, facturacion)

plt.title("Facturación total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Facturación")

plt.show()
```

# 2. Ventas por categoría

```python
ventas_categoria_t1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria_t2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_t3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_t4 = tienda4["Categoría del Producto"].value_counts()

print("Ventas por categoría - Tienda 1")
print(ventas_categoria_t1)

print("\nVentas por categoría - Tienda 2")
print(ventas_categoria_t2)

print("\nVentas por categoría - Tienda 3")
print(ventas_categoria_t3)

print("\nVentas por categoría - Tienda 4")
print(ventas_categoria_t4)
```

```python
ventas_categoria = tienda["Categoría del Producto"].value_counts()

plt.pie(
    ventas_categoria.values,
    labels=ventas_categoria.index,
    autopct='%1.1f%%'
)

plt.title("Distribución de ventas por categoría - Tienda 1")

plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda1 = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()
```

```python

calificaciones = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Calificación promedio": [
        calificacion_tienda1,
        calificacion_tienda2,
        calificacion_tienda3,
        calificacion_tienda4
    ]
}

df_calificaciones = pd.DataFrame(calificaciones)

df_calificaciones
```

```python
calificaciones = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.plot(tiendas, calificaciones, marker="o")

plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")

plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_productos_t1 = tienda["Producto"].value_counts()
ventas_productos_t2 = tienda2["Producto"].value_counts()
ventas_productos_t3 = tienda3["Producto"].value_counts()
ventas_productos_t4 = tienda4["Producto"].value_counts()

print("Más vendido Tienda 1:", ventas_productos_t1.idxmax())
print("Menos vendido Tienda 1:", ventas_productos_t1.idxmin())

print("Más vendido Tienda 2:", ventas_productos_t2.idxmax())
print("Menos vendido Tienda 2:", ventas_productos_t2.idxmin())

print("Más vendido Tienda 3:", ventas_productos_t3.idxmax())
print("Menos vendido Tienda 3:", ventas_productos_t3.idxmin())

print("Más vendido Tienda 4:", ventas_productos_t4.idxmax())
print("Menos vendido Tienda 4:", ventas_productos_t4.idxmin())
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda1 = tienda["Costo de envío"].mean()
envio_promedio_tienda2 = tienda2["Costo de envío"].mean()
envio_promedio_tienda3 = tienda3["Costo de envío"].mean()
envio_promedio_tienda4 = tienda4["Costo de envío"].mean()
```

```python
envios = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo de envío promedio": [
        envio_promedio_tienda1,
        envio_promedio_tienda2,
        envio_promedio_tienda3,
        envio_promedio_tienda4
    ]
}

df_envios = pd.DataFrame(envios)

df_envios
```

# **Informe Final**

## **Indroducción**

El objetivo de este análisis fue evaluar el rendimiento de las cuatro tiendas de la cadena Alura Store con el fin de recomendar cuál de ellas debería vender el Sr. Juan para iniciar un nuevo emprendimiento. Para ello, se analizaron diferentes métricas relevantes como los ingresos totales, las categorías de productos más vendidas, las calificaciones promedio de los clientes, los productos más y menos vendidos y el costo de envío promedio.

A partir de estos indicadores y las visualizaciones generadas, fue posible identificar patrones de comportamiento en las ventas y en la satisfacción de los clientes.

##**Desarrollo del análisis**

En primer lugar, se analizó la facturación total de cada tienda, lo que permitió identificar cuáles generan mayores ingresos. Este indicador es clave para evaluar el rendimiento financiero de cada tienda.

Posteriormente, se examinaron las ventas por categoría, lo que permitió identificar los tipos de productos más populares entre los clientes y aquellos con menor demanda. Este análisis ayuda a comprender el perfil de consumo en cada tienda.

También se evaluó la calificación promedio de los clientes, lo que permitió medir el nivel de satisfacción de los compradores con respecto a los productos y servicios ofrecidos por cada tienda.

Adicionalmente, se identificaron los productos más vendidos y menos vendidos, lo cual proporciona información valiosa sobre el rendimiento individual de los productos dentro del catálogo de cada tienda.

Finalmente, se analizó el costo de envío promedio, un factor importante ya que el cliente es quien asume este costo, y un valor elevado podría afectar negativamente la decisión de compra.

Todos estos resultados fueron representados mediante gráficos que facilitaron la interpretación visual de los datos y permitieron comparar el desempeño entre las diferentes tiendas.

## **Conclusiones**

Con base en los análisis realizados, se recomienda que el Sr. Juan considere vender la tienda que presenta menor facturación, menor satisfacción de los clientes y menor rendimiento en ventas en comparación con las demás tiendas. Estos factores indican que dicha tienda tiene un desempeño inferior dentro de la cadena.

Al vender la tienda con menor rendimiento, el Sr. Juan podrá concentrar sus recursos y esfuerzos en fortalecer las tiendas más rentables y con mejores indicadores de satisfacción del cliente.

Este análisis demuestra cómo el uso de herramientas de análisis de datos y visualización permite tomar decisiones estratégicas basadas en evidencia, facilitando la identificación de oportunidades de mejora dentro de un negocio.
